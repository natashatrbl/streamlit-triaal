## SCATTER PLOT

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv(r"D:\PROJECT TASHA\PYTHON\learning\streamlit\data\iris.csv")

print(df.columns)

plot = px.scatter(data_frame=df, 
                  size= 'sepal_width',
                  x='sepal_length', 
                  color='species',
                  y='petal_length',
                  template='plotly_dark', 
                  title='Plot of Sepal Length and Petal Length')

plot.show()
```

## Bar Chart

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv(r"D:\PROJECT TASHA\PYTHON\learning\streamlit\data\iris.csv")

print(df.head())

df1 = df.groupby(['species']).mean().reset_index()

print(df1.head())

plot = px.bar(data_frame= df1,
              x='species',
              y='petal_width',
              title='Petal Width Across Species')

plot.show()
```

## Box Plot & Violin Plot

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv(r"D:\PROJECT TASHA\PYTHON\learning\streamlit\data\iris.csv")

print(df.columns)

plot = px.box(data_frame=df,
              x='species',
              y='sepal_width', 
              color='species',
              title='Sepal Width Across Different Species')

plot.show()

plot1 = px.violin(
    data_frame=df,
              x='species',
              y='sepal_width', 
              color='species',
              box=True,
              title='Sepal Width Across Different Species'
)

plot1.show()
```

## HISTOGRAMS

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv(r"D:\PROJECT TASHA\PYTHON\learning\streamlit\data\iris.csv")

print(df.columns)

plot = px.histogram(data_frame=df,
                    x='sepal_length',
                    nbins=8,
                    color='species',
                    barmode='overlay',
                    template='plotly_dark',
                    title='Distribution of Sepal Length')

plot.show()
```

## MAPBOX PLOT

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv(r"D:\PROJECT TASHA\PYTHON\learning\streamlit\data\us-cities-top-1k.csv")

plot = px.scatter_mapbox(
    data_frame=df, 
    lat='lat', 
    lon='lon', 
    size='Population',
    zoom=4,
    title='Population of US Cities',
    mapbox_style='open-street-map', 
)

plot.show()
```

## PIE CHART

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv(r"D:\PROJECT TASHA\PYTHON\learning\streamlit\data\tips.csv")

print(df.columns)

plot = px.pie(
    data_frame=df,
    values='tip', 
    names='sex', 
    facet_col='smoker',
    hole=0.4,
    title='Percentage of Tip from Each Gender'
)

plot.show()
```
